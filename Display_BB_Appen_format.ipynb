```python
import cv2
import urllib.request
import numpy as np
import pandas as pd
import json
```

```python
df = pd.read_csv('ocr_multi_bb_example.csv', encoding = 'UTF-8')
df
```

```python
for index, row in df.iterrows():
    # read the image from url
    resp = urllib.request.urlopen(row['google_url'])
    img = np.asarray(bytearray(resp.read()), dtype="uint8")
    img = cv2.imdecode(img, cv2.IMREAD_COLOR)
    img_copy = img.copy()
 
    # read the column of annotation as a list of dictionaries
    bb = json.loads(row['annotation'])
    
    # since every image can have more than one BB we need to check how many objects were marked on it and draw all the BB
    try:
        for i in range(len(bb)):
            bb_dict = dict(bb[i])
            bb_coor = dict(bb_dict['coordinates'])
            x = bb_coor['x']
            y = bb_coor['y']
            w = bb_coor['w']
            h = bb_coor['h']
            x, y, w, h = bb_coor['x'], bb_coor['y'], bb_coor['w'], bb_coor['h']
            cv2.rectangle(img_copy, (x, y), (x+w, y+h), (0, 255, 0), 2)    
            
             # display the image
            cv2.imshow('Image with bounding box', img_copy)
            cv2.waitKey(0)
    except:
        print("Error in line ", row['row_id']," ", row['annotation'])
        
    
    
    
# close all windows
cv2.destroyAllWindows()
```
